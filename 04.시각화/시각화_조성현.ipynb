```python
# 이벤트 유형별 고유 사용자 수 계산
event_types = ['view', 'cart', 'purchase']
user_counts = {event: df[df['event_type'] == event]['user_id'].nunique() for event in event_types}

# 데이터 준비
event_labels = list(user_counts.keys())
user_count_values = list(user_counts.values())

# 막대그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(event_labels, user_count_values, color=['#66b3ff', '#99ff99', '#ff9999'])
plt.xlabel('Event Type')
plt.ylabel('Number of Unique Users')
plt.title('Number of Unique Users by Event Type')
plt.show()
```

```python
# 함수 정의: 첫 활동 이후 재방문 여부 계산
def calculate_revisit(df, event_type):
    first_event_df = df[df['event_type'] == event_type].drop_duplicates(subset=['user_id'])
    all_events_df = df[df['event_type'] == event_type]
    
    repeated_events = all_events_df[all_events_df.duplicated(subset=['user_id'], keep=False)]
    first_time_events = first_event_df[~first_event_df['user_id'].isin(repeated_events['user_id'])]
    
    repeated_event_users = repeated_events['user_id'].nunique()
    first_time_event_users = first_time_events['user_id'].nunique()
    
    return repeated_event_users, first_time_event_users

# 구매 이벤트 계산
repeated_purchase_users, first_time_purchase_users = calculate_revisit(df, 'purchase')

# 조회 이벤트 계산
repeated_view_users, first_time_view_users = calculate_revisit(df, 'view')

# 데이터 준비
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# 구매 이벤트 원형 차트
labels_purchase = ['Repeated Purchase', 'First Time Purchase Only']
sizes_purchase = [repeated_purchase_users, first_time_purchase_users]
colors_purchase = ['#ff9999', '#66b3ff']
explode_purchase = (0.1, 0)  # 첫 번째 조각을 약간 분리

axs[0].pie(sizes_purchase, explode=explode_purchase, labels=labels_purchase, colors=colors_purchase,
           autopct='%1.1f%%', shadow=True, startangle=90)
axs[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[0].set_title('Percentage of Users Who Made Repeated Purchases')

# 조회 이벤트 원형 차트
labels_view = ['Repeated View', 'First Time View Only']
sizes_view = [repeated_view_users, first_time_view_users]
colors_view = ['#99ff99', '#ffcc99']
explode_view = (0.1, 0)  # 첫 번째 조각을 약간 분리

axs[1].pie(sizes_view, explode=explode_view, labels=labels_view, colors=colors_view,
           autopct='%1.1f%%', shadow=True, startangle=90)
axs[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[1].set_title('Percentage of Users Who Made Repeated Views')

# 차트 보여주기
plt.tight_layout()
plt.show()
```

```python
# 각 이벤트별 첫 활동 완료 여부 계산 함수
def calculate_first_activity(df, event_type):
    first_activity_df = df[df['event_type'] == event_type].drop_duplicates(subset=['user_id'])
    total_users = df['user_id'].nunique()
    first_activity_users = first_activity_df['user_id'].nunique()
    no_first_activity_users = total_users - first_activity_users
    
    return first_activity_users, no_first_activity_users

# 각 이벤트별 계산 및 시각화
event_types = ['purchase', 'cart']
colors = ['#ff9999', '#66b3ff', '#99ff99']

# 서브플롯 생성
fig, axs = plt.subplots(1, len(event_types), figsize=(14, 7))

for i, event_type in enumerate(event_types):
    first_activity_users, no_first_activity_users = calculate_first_activity(df, event_type)
    
    labels = [f'First {event_type.capitalize()}', f'No {event_type.capitalize()}']
    sizes = [first_activity_users, no_first_activity_users]
    explode = (0.1, 0)  # 첫 번째 조각을 약간 분리
    
    # 원형 차트 그리기
    axs[i].pie(sizes, explode=explode, labels=labels, colors=colors[i:], autopct='%1.1f%%',
               shadow=True, startangle=90)
    axs[i].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    # 제목 추가
    axs[i].set_title(f'Percentage of Users Who Completed First {event_type.capitalize()}')

# 차트 간격 조정 및 보여주기
plt.tight_layout()
plt.show()
```
